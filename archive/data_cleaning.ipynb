```python
# Check if Wochentag and tag_nr are always matching (1 Montag, 2 Dienstag, 3 Mittwoch, 4 Donnerstag, 5 Freitag, 6 Samstag, 7 Sonntag)
day_mapping = {
    1: 'Montag',
    2: 'Dienstag',
    3: 'Mittwoch',
    4: 'Donnerstag',
    5: 'Freitag',
    6: 'Samstag',
    7: 'Sonntag'
}

data_1_drop['is_matching'] = data_1_drop.apply(lambda row: day_mapping[row['tag_nr']] == row['Wochentag'], axis=1)
data_2_drop['is_matching'] = data_2_drop.apply(lambda row: day_mapping[row['tag_nr']] == row['Wochentag'], axis=1)

all_match = data_1_drop['is_matching'].all() and data_2_drop['is_matching'].all()
print("Do all rows match?", all_match)
```

```python
import pandas as pd

# Copy the dataset
df_1 = data_1_short.copy()
df_2 = data_2_short.copy()

def change_date_format(df):
    # Split 'Gemessen am' column by '+'
    df['Gemessen am'] = df['Gemessen am'].str.split('+').str[0]

    # Convert 'Gemessen am' to datetime
    df.loc[:, 'Gemessen_am_datetime'] = pd.to_datetime(df['Gemessen am'], errors='coerce').dt.floor('S')
    df.loc[:, 'Gemessen_am_date'] = df['Gemessen_am_datetime'].dt.date
    df.loc[:, 'Gemessen_am_time'] = df['Gemessen_am_datetime'].dt.time

    # Update 'datum_tag' with the corrected 'Gemessen_am_date'
    df.loc[:, 'is_date_matching'] = df['Gemessen_am_date'].astype(str) == df['datum_tag']

    # Recalculate 'tag_nr' (day of the week, 1=Monday, 7=Sunday)
    df['Gemessen_am_tag_nr'] = df['Gemessen_am_datetime'].dt.dayofweek +1

    # Relable 'Wochentag'
    df['Gemessen_am_Arbeitstag'] = df['Gemessen_am_tag_nr'].apply(lambda x: 'Werktage' if x <= 5 else 'Wochenende')

    return df

df_1_changed = change_date_format(df_1)
df_2_changed = change_date_format(df_2)

df_2_changed
```

```python
clumns_to_drop = ['Gemessen am', 'tag_nr', 'datum_tag', 'Arbeitstag', 'is_matching', 'is_date_matching']
columns_rename={
    'Gemessen_am_datetime': 'Measurement_Timestamp',
    'Gemessen_am_date': 'Measurement_Date',
    'Gemessen_am_time': 'Measurement_Time',
    'Gemessen_am_tag_nr': 'Day_Number',
    'Gemessen_am_Arbeitstag': 'Workday_Type',
    'Passanten in Richtung Neumarkt': 'Pedestrians_to_Neumarkt',
    'Passanten in Richtung Multergasse': 'Pedestrians_to_Multergasse',
    'Passanten in beide Richtungen (Summe)' : 'Pedestrians_Total'
}


cleaned_df_1 = df_1_changed.drop(columns=clumns_to_drop).rename(columns=columns_rename)
cleaned_df_2 = df_2_changed.drop(columns=clumns_to_drop).rename(columns=columns_rename)

# Specify the new column order
new_column_order = ['Measurement_Timestamp', 'Measurement_Date', 'Measurement_Time', 'Day_Number', 'Workday_Type', 'Pedestrians_to_Neumarkt', 'Pedestrians_to_Multergasse', 'Pedestrians_Total']  # Desired order of columns

cleaned_1 = cleaned_df_1[new_column_order]
cleaned_2 = cleaned_df_2[new_column_order]
```

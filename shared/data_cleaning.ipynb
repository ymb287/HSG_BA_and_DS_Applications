## Remove columns

```python
# Cehck if both have the same columns
print(data_1.columns)
print(data_2.columns)
```

```python
# Check for nan values
print(data_1.isnull().sum())
print(data_2.isnull().sum())
```

```python
# Get unique values of the columns Standort, Sensor ID, Sensorname, device_name and geo_point_2d
for col in ['Standort', 'Sensor ID', 'Sensorname', 'geo_point_2d']:
    print(data_1[col].unique())
    print(data_2[col].unique())

```

```python
# Drop those columns
data_1_drop = data_1.drop(columns=['Standort', 'Sensor ID', 'Sensorname', 'geo_point_2d'])
data_2_drop = data_2.drop(columns=['Standort', 'Sensor ID', 'Sensorname', 'geo_point_2d', 'device_name'])
```

```python
# Check if Wochentag and tag_nr are always matching (1 Montag, 2 Dienstag, 3 Mittwoch, 4 Donnerstag, 5 Freitag, 6 Samstag, 7 Sonntag)
day_mapping = {
    1: 'Montag',
    2: 'Dienstag',
    3: 'Mittwoch',
    4: 'Donnerstag',
    5: 'Freitag',
    6: 'Samstag',
    7: 'Sonntag'
}

data_1_drop['is_matching'] = data_1_drop.apply(lambda row: day_mapping[row['tag_nr']] == row['Wochentag'], axis=1)
data_2_drop['is_matching'] = data_2_drop.apply(lambda row: day_mapping[row['tag_nr']] == row['Wochentag'], axis=1)

all_match = data_1_drop['is_matching'].all() and data_2_drop['is_matching'].all()
print("Do all rows match?", all_match)
```

```python
data_1_short = data_1_drop.drop(columns=['Wochentag'])
data_2_short = data_2_drop.drop(columns=['Wochentag'])
```

```python
# Print columns
print(data_1_short.columns)
print(data_2_short.columns)
```

## Check for Date missmatches

```python
df = data_1_short.copy()

df['Gemessen am'] = df['Gemessen am'].str.split('+').str[0]
df.loc[:, 'Gemessen_am_datetime'] = pd.to_datetime(df['Gemessen am'], errors='coerce')
df['Gemessen_am_date'] = df['Gemessen_am_datetime'].dt.date
df.loc[:, 'is_date_matching'] = df['Gemessen_am_date'].astype(str) == df['datum_tag']


# Filter rows where the date is not matching
mismatched_rows = df.loc[~df['is_date_matching']]

# Display mismatched rows
mismatched_rows
```
